```python
from functools import reduce

def vector_add(a : list, b : list) -> list:
    return [u + v for u, v in zip(a, b)]

def vector_subtract(a : list, b : list) -> list:
    return [u - v for u, v in zip(a, b)]

def vector_sum(vectors : list) -> list:
    return reduce(vector_add, vectors)

def scalar_multiply(f : float, v : list) -> list:
    return [f * ii for ii in v]

print(vector_add([2,3], [7,8]))
print(vector_subtract([2,3], [7,8]))
print(scalar_multiply(3, vector_sum([[2,3], [7,8],[1,2]])))
```
