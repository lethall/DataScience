1) Consider the star Aldebaran. At some point in time, it's horizontal coordinates in Glasgow (latitude = 55.861 degrees North) are: A = 282.15 degrees, h = 28.05 degrees; calculate the hour coordinates of this star (H , delta).

```python
latitude = 55.861
A = 282.15
h = 28.05
hra = Spherical().law_of_cosine(latitude, 90-h, 180-A)
decl = Spherical().law_of_sine_SSA(hra, latitude, 180-A)
print(f"Aldebaran RA: {hra/15:.2f}h decl: {decl:.3f}\xb0")
# Stellarium says 4h35m55.5s 16d30'27.1"
```

```python
latitude = 55.861
A = 282.15
h = 28.05
decl, hra = position(latitude, h, A)
ra = DMS(hra/15)
print(f"Aldebaran RA: {ra.deg:.0f}:{ra.min:2.0f}:{ra.sec:02.2f} ({hra:.5f}\xb0) decl: {DMS(decl)}")
```

2) Now, let's take the star Castor. At some point in time, it's hour coordinates in Glasgow are: H = 1.858 hours, delta = 31.83 degrees; calculate the horizontal coordinates at that time (A , h)

```python
latitude = 55.861
hra = 1.858 * 15
decl = 31.83
h, azm = position(latitude, decl, hra, False)
print(f"Castor azm: {DMS(azm)} alt: {DMS(h)}")
```
