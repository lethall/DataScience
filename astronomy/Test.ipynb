```python
from ipycanvas import Canvas

canvas = Canvas(width=200, height=200)

canvas.fill_rect(25, 25, 100, 100)
canvas.clear_rect(45, 45, 60, 60)
canvas.stroke_style = "red"
canvas.stroke_rect(50, 50, 50, 50)

canvas
```
