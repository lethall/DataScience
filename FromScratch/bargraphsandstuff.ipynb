```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()], histogram.values(), 10, edgecolor=(0,0,0))

plt.axis([-5, 105, 0, 5])

plt.title("Distribution of Exam grades")
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("Number of Students")
plt.show
```

It is imporant to note how this is visually constructed, with the scores between each decile marked together - "Counter(min(grade // 10 * 10, 90) for grade in grades)
